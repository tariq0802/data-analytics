# Central limit theoram
### What is Central Limit Theorem?
* The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution.
* This theorem enables the use of normal distributions in statistical inference and hypothesis testing.

### Assumptions of Central Limit Theorem -
* Data should be randomly sampled.
* Samples should be collected from the same distribution.
* Samples should be independent of each other.
* Samples should have sufficient size (>=30).

### Sampling Distribution of Mean
Sampling distribution of mean is simply the distribution of sample means over repeated samples.
* Original uniform distribution
* Sample mean of 10 over 100 samples
* Sample mean of 20 over 100 samples

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate a population of data
population = np.random.normal(loc=50, scale=10, size=10000)

# Select random samples from the population
sample_sizes = [10, 30, 50, 100]
num_samples = 1000

# Calculate the sample means
sample_means = []

for size in sample_sizes:
    means = []
    
    for _ in range(num_samples):
        sample = np.random.choice(population, size=size, replace=False)
        mean = np.mean(sample)
        means.append(mean)
        
    sample_means.append(means)
    
# Creating distribution plot for samples
fig, axs = plt.subplots(2, 2, figsize=(15,10))
axs = axs.flatten()
sns.set_style('darkgrid')

for i, ax in enumerate(axs):
    sns.histplot(x=sample_means[i], ax=ax)
    ax.set_xlabel('Sample Mean')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Sample Size: {sample_sizes[i]}', fontsize=16, fontweight='bold')
    
plt.tight_layout()
plt.show()
```
