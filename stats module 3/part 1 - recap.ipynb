# Lesson 1: Central Tendency and Measures of Dispersion
## Central Tendencies
* Central Tendency:
Evaluate a point estimate for the data
* Mean:
Average value in the data
* Median:
Middle value in the data
* Mode:
Value with highest frequency in the data

## Measures of Variability
* Variance is average degree to which each data point differs from the mean.
* Standard Deviation is simply the square root of variance.
* Range is difference between max and min of the data.
* Quartile simply divides the observations in data into 4 equal intervals or quarters which are more or less of equal size.
* Interquartile range is difference between 75th percentile and 25th percentile.

Outliers are set of extreme values that stand significantly different from other observations in a dataset. These can be caused by data entry errors, measurement errors, collection of sample errors and natural variation.

```python
df = {'Employee ID': ['E01', 'E02', 'ะ03', 'ะ04', 'ะ05', 'ะ06', 'ะ07'],
    'Age': [30,53,42,29,95,36,52],
    'Salary': [0.781571617,0.2484892,2.997425024,0.370216911,0.574881814,0.930060076,0.643781331],
    'Years of Experience': [5,50,14,18,35,10,24]}
df = pd.DataFrame(df)
df
```

## Detecting outliers with boxplots

```python
# Creating subplots for 3 columns
fig, axs = plt.subplots(1,3, figsize=(16,6))

# columns for box plots
columns = ['Age', 'Salary', 'Years of Experience']

# Boxplot
for i, column in enumerate(columns):
    sns.boxplot(y=df[column], ax=axs[i])
    axs[i].set_title(column, fontsize=16, fontweight='bold')
```

### Outcome from Boxplots
* in Age column there is a value greater than 90. which is very rare case. so it might be outlier
* in Salary column there is a value around 3. which is very far from range and it might be outlier
* There is not any outlier presen in Years of Experience column

### Detect outliers with 3 methods

```python
# Percentile method
def percentile_method(data, col):
    lower = data[col].quantile(0.01)
    upper = data[col].quantile(0.99)
    outliers = data[(data[col] < lower) | (data[col] > upper)]
    return outliers

percentile_method(df, 'Age')
```
