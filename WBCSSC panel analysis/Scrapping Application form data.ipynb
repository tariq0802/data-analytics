## SCRAPPING APPLICATION FORM DATA

### GET FORM LINKS SUBJECT WISE

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import selenium.webdriver.common.keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from datetime import datetime
import pandas as pd
import time
import os
```

```python
service = Service(executable_path='/Applications/Chdriver/chromedriver')
driver = webdriver.Chrome(service=service)

url = 'http://westbengalssc.com/sscorg/wbssc/slstAppReprintUpperPrimaryPT/'

subjects = ["ARABIC(UP)","BENGALI(UP)", "BIO. SCIENCE(UP)", "ENGLISH(UP)", "GEOGRAPHY(UP)", 
       "HINDI(UP)", "HISTORY(UP)", "NEPALI(UP)", "ORIYA(UP)", "PERSIAN(UP)", "PURE SCIENCE(UP)", 
       "SANSKRIT(UP)", "SANTHALI(UP)", "TELEGU(UP)", "URDU(UP)"]
mediums = ["BENGALI", "HINDI", "ENGLISH", "NEPALI", "URDU", "TELEGU", "SANTHALI"]
categories = ["GEN", "OBC-A", "OBC-B", "PH-HI", "PH-OH", "PH-VH", "SC", "ST"]
genders = ["MALE", "FEMALE"]

driver.get(url)
time.sleep(1)

data = []

for sub in subjects:
    try:
        sub_field = Select(driver.find_element(By.ID, 'subject'))
        sub_field.select_by_value(sub)
        time.sleep(1)
    except NoSuchElementException:
        continue
          
    for medium in mediums:
        try:
            medium_field = Select(driver.find_element(By.ID, 'medium'))
            medium_field.select_by_value(medium)
            time.sleep(1)
        except NoSuchElementException:
            continue

        for category in categories:
            try:
                category_field = Select(driver.find_element(By.ID, 'category'))
                category_field.select_by_value(category)
                time.sleep(1)
            except NoSuchElementException:
                continue                 

            for gender in genders:
                try:
                    gender_field = Select(driver.find_element(By.ID, 'gender'))
                    gender_field.select_by_value(gender)
                    time.sleep(1)
                except NoSuchElementException:
                    continue

                try:
                    submit_button = driver.find_element(By.ID, "VacBtnId")
                    submit_button.click()
                    time.sleep(1)
                except NoSuchElementException:
                    continue

                rows = driver.find_elements(By.XPATH, '//tbody/tr')[:-1]

                for row in rows:
                    serial = row.find_element(By.CSS_SELECTOR, 'td:nth-child(1)').text
                    name = row.find_element(By.CSS_SELECTOR, 'td:nth-child(2)').text
                    roll = row.find_element(By.CSS_SELECTOR, 'td:nth-child(3)').text
                    link = row.find_element(By.CSS_SELECTOR, 'td:nth-child(3) a').get_attribute('href')

                    data.append({'Serial': serial, 'Name': name, 'Roll': roll, 'Link': link})

                driver.back()
                time.sleep(1)

driver.quit()
```

```python
form_links = pd.DataFrame(data)
```

```python
service = Service(executable_path='/Applications/Chdriver/chromedriver')
driver = webdriver.Chrome(service=service)

for link in form_links:
    driver.get(link)
    
    name = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[1]/td[2]').text
    id_text = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[1]/td[3]/div/strong').text
    parts = id_text.split('\n')
    app_id = parts[0].split(' :')[1]
    roll_no = parts[1].split(' :')[1]
    gender = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[8]/td[2]').text.strip()
    care_vill = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[2]/td[2]')
    father = care_vill[0].text
    address1 = care_vill[1].text
    dob_add2 = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[3]/td[2]')
    dob = dob_add2[0].text
    address2 = dob_add2[1].text
    med_dis = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[4]/td[2]')
    medium = med_dis[0].text
    dist = med_dis[1].text
    pin = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[5]/td[2]')[1].text
    gr_ph = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[6]/td[2]')
    group = gr_ph[0].text
    phone = gr_ph[1].text
    sub_mail = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[7]/td[2]')
    subject = sub_mail[0].text
    mail = sub_mail[1].text
    gender = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[8]/td[2]').text
    category = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[9]/td[2]').text
    mp_marks = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[11]/td/table/tbody/tr[2]/td')
    mp_full = mp_marks[1].text
    mp_obtd = mp_marks[2].text
    mp_per = mp_marks[3].text
    hs_marks = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[11]/td/table/tbody/tr[3]/td')
    hs_full = hs_marks[1].text
    hs_obtd = hs_marks[2].text
    hs_per = hs_marks[3].text
    grad_marks = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[11]/td/table/tbody/tr[5]/td')
    grad_full = grad_marks[1].text
    grad_obtd = grad_marks[2].text
    grad_per = grad_marks[3].text
    tr_marks = driver.find_elements(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[11]/td/table/tbody/tr[6]/td')
    tr_full = tr_marks[1].text
    tr_obtd = tr_marks[2].text
    tr_per = tr_marks[3].text
    tet_per = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[12]/td[2]').text
    training = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[13]/td[2]').text
    scale = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[14]/td[2]').text
    fees = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[15]/td[2]').text
    religion = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[16]/td[2]').text
    submitted = driver.find_element(By.XPATH, '//*[@id="rounded-corner"]/tbody/tr[19]/td').text
    submit_date = submitted.split(': ')[1]
        
    
    data.append({"Application ID": app_id, "Roll No": roll_no, "Name": name, 
                 "Parent's Name": father, "Date of Birth": dob, "Medium": medium, 
                 "Group": group, "Subject": subject, "Gender": gender, "Category": category, 
                 "MP FM": mp_full, "MP Obtained": mp_obtd, "MP Percentage": mp_per,
                 "HS FM": hs_full, "HS Obtained": hs_obtd, "HS Percentage": hs_per,
                 "Grad FM": grad_full, "Grad Obtained": grad_obtd, "Grad Percentage": grad_per,
                 "Training FM": tr_full, "Training Obtained": tr_obtd, "Training Percentage": tr_per,
                 "TET Percentage": tet_per, "Training": training, "Present Scale": scale, 
                 "Fees Paid": fees, "Religion": religion, "Submitted On": submit_date, "Address 1": address1, 
                 "Address 2": address2, "District": dist, "PIN": pin, "Phone No": phone, "Email": mail})
    
driver.quit()

df = pd.DataFrame(data)
df.to_csv("data/form_data.csv", index=False)
```
