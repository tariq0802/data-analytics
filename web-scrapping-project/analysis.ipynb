```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = [
    'Mohammad Rizwan', 'Dawid Malan', 'Reeza Hendricks', 'Yasir Khan', 'Iftikhar Ahmed', 'Khushdil Shah', 
    'David Willey', 'Usama Mir', 'Abbas Afridi', 'Mohammad Ali', 'Shahnawaz Dahani', 'Saim Ayub', 'Babar Azam', 
    'Mohammad Haris', 'Haseeb Ullah Khan', 'Rovman Powell', 'Asif Ali', 'Paul Walter', 'Luke Wood', 'Arif Yaqoob', 
    'Naveen Ul Haq', 'Salman Irshad']
player_df = pd.read_csv('players.csv')
df = player_df[player_df['Name'].isin(names)]
df = df[['Name', 'Date', 'Runs', 'Wicket']]
df['Date'] = pd.to_datetime(df['Date'])
df.fillna(0, inplace=True)
df.sort_values(by=['Name', 'Date'], ascending=[True, True], inplace=True)
df
```

```python
pd.set_option('display.max_rows', None)
df['Cum_runs'] = df.groupby('Name')['Runs'].cumsum()
df['Cum_Dis'] = df.groupby('Name').cumcount() + 1
df['Cum_Avg'] = round(df['Cum_runs'] / df['Cum_Dis'], 2)

df['Cum_Wickets'] = df.groupby('Name')['Wicket'].cumsum()
df['Cum_W_Avg'] = round(df['Cum_Wickets'] / df['Cum_Dis'], 2)
df
```

```python
df1 = df[df['Cum_Avg'] > 10]
df2 = df[df['Cum_W_Avg'] > 0.5]
sns.set_style('darkgrid')

fig, axs = plt.subplots(2, 1, figsize=(16, 20))
sns.lineplot(x='Cum_Dis', y='Cum_Avg', hue='Name', data=df1, ax=axs[0], linewidth=3)
axs[0].set_title('Cumulative Runs Average')

sns.lineplot(x='Cum_Dis', y='Cum_W_Avg', hue='Name', data=df2, ax=axs[1], linewidth=3,)
axs[1].set_title('Cumulative wickets Average')
plt.show()
```
