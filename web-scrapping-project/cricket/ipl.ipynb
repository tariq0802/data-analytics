```python
import pandas as pd
import os

batter_files = []
bowler_files = []
for i in range(2008,2024):
    bat = f'ipl-{i}-batter.csv'
    bow = f'ipl-{i}-bowler.csv'
    batter_files.append(bat)
    bowler_files.append(bow)

batter_dfs = []
for file in batter_files:
    df = pd.read_csv(file)
    batter_dfs.append(df)
    
batters = pd.concat(batter_dfs, axis=0, ignore_index=True)

bowler_dfs = []
for file in bowler_files:
    df = pd.read_csv(file)
    bowler_dfs.append(df)
    
bowlers = pd.concat(bowler_dfs, axis=0, ignore_index=True)
bowlers
```

```python
batters['Match Date'] = pd.to_datetime(batters['Match Date'])
batters['Born'] = pd.to_datetime(batters['Born'])
bowlers['Match Date'] = pd.to_datetime(bowlers['Match Date'])
bowlers['Born'] = pd.to_datetime(bowlers['Born'])
print(batters.info())
print(bowlers.info())
```

```python
batter_file = 'ipl-batter.csv'
bowler_file = 'ipl-bowler.csv'

batters.to_csv(batter_file, index=False, header=not os.path.isfile(batter_file))
bowlers.to_csv(bowler_file, index=False, header=not os.path.isfile(bowler_file))
```
