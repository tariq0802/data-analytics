```python
import pandas as pd

batter_files = []
bowler_files = []
for i in range(2008,2024):
    bat = f'ipl-{i}-batter.csv'
    bow = f'ipl-{i}-bowler.csv'
    batter_files.append(bat)
    bowler_files.append(bow)

batter_dfs = []
for file in batter_files:
    df = pd.read_csv(file)
    batter_dfs.append(df)
    
batters = pd.concat(batter_dfs, axis=0, ignore_index=True)

bowler_dfs = []
for file in bowler_files:
    df = pd.read_csv(file)
    bowler_dfs.append(df)
    
bowlers = pd.concat(bowler_dfs, axis=0, ignore_index=True)
```

```python
batters['Date'] = pd.to_datetime(batters['Date'])
batters['Season'] = batters['Date'].dt.year
bowlers['Date'] = pd.to_datetime(bowlers['Date'])
bowlers['Season'] = bowlers['Date'].dt.year

batters
bowlers
```
